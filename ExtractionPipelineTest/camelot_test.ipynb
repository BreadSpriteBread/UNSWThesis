```python
import pandas as pd
import camelot
```

```python
pdf_file = "snb.pdf"  
```

```python
tables = camelot.read_pdf(pdf_file, pages='all', flavor='stream')

print(f"Extracted {len(tables)} tables from PDF")

balance_sheet_df = None

for i, table in enumerate(tables):
    df = table.df
    if df.apply(lambda row: row.astype(str).str.contains("Balance Sheet", case=False).any(), axis=1).any():
        balance_sheet_df = df
        print(f"Balance Sheet found in table {i}")
        break
```

```python
if balance_sheet_df is not None:
    balance_sheet_df = balance_sheet_df.dropna(axis=1, how='all')
    
    balance_sheet_df.rename(columns={balance_sheet_df.columns[0]: "Item"}, inplace=True)
    
    balance_sheet_df.reset_index(drop=True, inplace=True)
    
    print(balance_sheet_df.head())

    balance_sheet_df.to_csv("balance_sheet.csv", index=False)
    print("Balance sheet saved to balance_sheet.csv")
else:
    print("No balance sheet found in the document.")
```
